### checking for outliers

```python
all_countinous_columns = ['Age', 'DistanceFromHome', 'EmployeeCount',
       'MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike',
       'StandardHours', 'StockOptionLevel', 'TotalWorkingYears',
       'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']

for one_countinous_column in all_countinous_columns:
    print('Box Plot For ',one_countinous_column)
    plt.boxplot(df[one_countinous_column])
    plt.show()
```

### As we can see there is no outlier for: Age, Distace From Home, Employee Count, Percent Salary Hike, Standerd Hours

### Correlation with the target variable 

```python
df.corr()['Attrition_Yes']
```

```python
## sorting value to see the clear relation
correlation_attrition = dict(df.corr()['Attrition_Yes'])
correlation_attrition_sorted = sorted(correlation_attrition.items(),key=lambda item : item[1])

## converting the to DataFrame
correlation_attrition_df = pd.DataFrame(correlation_attrition_sorted)

## droping the last self row relation 
correlation_attrition_df = correlation_attrition_df.drop(35,axis = 0)

## dropign tha NA value
correlation_attrition_df.dropna(inplace=True)
correlation_attrition_df
```

### Lets Draw a chart to see the clear picture

```python
## settng config for plot
plt.rcParams['figure.figsize'] = [50, 20]
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : 35}

plt.rc('font', **font)
```

```python
plt.xticks(rotation='vertical')
plt.bar(correlation_attrition_df[0],correlation_attrition_df[1])
```
